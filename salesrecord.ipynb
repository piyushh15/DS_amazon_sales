# **About Data**

columns -

    Region: The geographical region where the sale or order was made.
    Country: The country where the sale or order was made.
    Item Type: The type of item sold (e.g., electronics, clothing, etc.).
    Sales Channel: The channel through which the sale was made (e.g., online, retail).
    Order Priority: The priority level of the order (e.g., high, medium, low).
    Order Date: The date when the order was placed.
    Order ID: A unique identifier for the order.
    Ship Date: The date when the order was shipped.
    Units Sold: The number of units sold in the order.
    Unit Price: The price per unit of the item.
    Unit Cost: The cost per unit of the item.
    Total Revenue: The total revenue from the order (Units Sold * Unit Price).
    Total Cost: The total cost for the order (Units Sold * Unit Cost).
    Total Profit: The total profit from the order (Total Revenue - Total Cost).
    Order Year: The year when the order was placed.
    Order Month: The month when the order was placed.
    Order Weekday: The day of the week when the order was placed.
    Unit Margin: The profit margin per unit (Unit Price - Unit Cost).
    Order_Ship_Days: The number of days between the order date and the ship date.

## **Product year Revenue**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def year_type_region_query(a1, ItemType, Orderyear):
    revenue = []
    year = []
    item_type = []
    for i in ItemType:
        for j in Orderyear:
            val = sum(a1[(a1['Item Type'] == i) & (a1['Order year'] == j)]['Total Revenue'].values)
            revenue.append(val)
            year.append(j)
            item_type.append(i)
    return revenue, year, item_type

revenue, year, item_type = year_type_region_query(a1, ItemType, Orderyear)
revenue_millions = [r / 1e6 for r in revenue]
df_temp = pd.DataFrame({
    'Year': year,
    'ItemType': item_type,
    'Revenue (Millions)': revenue_millions
})
plt.figure(figsize=(12, 6))
sns.barplot(data=df_temp, x='Year', y='Revenue (Millions)', hue='ItemType')
plt.title('Total Revenue by Item Type and Year (in Millions) Asia')
plt.xlabel('Year')
plt.ylabel('Total Revenue (Millions)')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def year_type_region_query(a1, ItemType, Orderyear):
    revenue = []
    year = []
    item_type = []
    for i in ItemType:
        for j in Orderyear:
            val = sum(a1[(a1['Item Type'] == i) & (a1['Order year'] == j)]['Total Revenue'].values)
            revenue.append(val)
            year.append(j)
            item_type.append(i)
    return revenue, year, item_type

def analytics_region(strr, data_frame):
    temp = data_frame[data_frame['Region'] == strr]
    ItemType = list(temp['Item Type'].value_counts().index)
    Orderyear = list(temp['Order year'].value_counts().index)
    return year_type_region_query(temp, ItemType, Orderyear)

def every_region(data_frame):
    regions = list(data_frame['Region'].value_counts().index)
    n_regions = len(regions)
    fig, axes = plt.subplots(nrows=n_regions, ncols=1, figsize=(12, n_regions * 6), sharex=True)

    if n_regions == 1:
        axes = [axes]

    for ax, region in zip(axes, regions):
        revenue, year, item_type = analytics_region(region, data_frame)
        df = pd.DataFrame({
            'Year': year,
            'ItemType': item_type,
            'Revenue': revenue
        })
        df['Revenue (Millions)'] = df['Revenue'] / 1e6

        sns.barplot(data=df, x='Year', y='Revenue (Millions)', hue='ItemType', ax=ax)
        ax.set_title(f'Total Revenue by Year for {region}')
        ax.set_xlabel('Year')
        ax.set_ylabel('Revenue (Millions)')
        ax.legend(title='Item Type', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.suptitle('Total Revenue by Year for Each Region (in Millions)', y=1.02)
    plt.show()


every_region(df)
```

```python
def pie_chart_year_wise(year, data_frame):

    year_data = data_frame[data_frame['Year'] == year]
    grouped_data = year_data.groupby('ItemType')['Revenue (Millions)'].sum().reset_index()
    fig, ax = plt.subplots()
    ax.pie(grouped_data['Revenue (Millions)'], labels=grouped_data['ItemType'], autopct='%1.1f%%', startangle=90)
    ax.axis('equal')
    ax.set_title(f'Revenue Distribution by Item Type for Year {year}')

    plt.show()
unique_years = sorted(df_temp['Year'].unique())

for year in unique_years:
    pie_chart_year_wise(year, df_temp)
```

```python
import pandas as pd
import plotly.graph_objects as go

# Assuming df is already defined
regions = list(df['Region'].value_counts().index)
years = list(df['Order year'].value_counts().index)
typeoff = list(df['Item Type'].value_counts().index)
revenues = []

def max_revenue_gen_product_per_area():
    for x in regions:
        for y in typeoff:
            for z in years:
                temp_df = df[(df['Region'] == x) & (df['Order year'] == z) & (df['Item Type'] == y)]
                val = temp_df['Total Revenue'].sum()
                revenues.append({'Region': x, 'Item Type': y, 'Year': z, 'Total Revenue': val})

max_revenue_gen_product_per_area()


revenue_df = pd.DataFrame(revenues)
revenue_df
```

```python
def max_revenue_item_per_region_per_year():
    result = []
    for year in revenue_df['Year'].unique():
        for region in revenue_df['Region'].unique():
            max_revenue_row = revenue_df[(revenue_df['Year'] == year) & (revenue_df['Region'] == region)].nlargest(1, 'Total Revenue')
            result.append(max_revenue_row)
    return pd.concat(result, ignore_index=True)

# Getting the result
max_items_df = max_revenue_item_per_region_per_year()
print(max_items_df)
```

```python
pivot_df = max_items_df.pivot(index='Year', columns='Region', values='Item Type')
fig = go.Figure(data=[go.Table(
    header=dict(values=['Year'] + list(pivot_df.columns),
                fill_color='paleturquoise',
                align='left',
                font=dict(size=12)),
    cells=dict(values=[pivot_df.index] + [pivot_df[col] for col in pivot_df.columns],
               fill_color='lavender',
               align='left',
               font=dict(size=10))
)])
fig.update_layout(
    title='Max Revenue Item per Region per Year',
    title_font=dict(size=18),
    autosize=False,
    width=1000,
    height=600,
    margin=dict(l=20, r=20, t=60, b=20),
)

fig.show()
```
